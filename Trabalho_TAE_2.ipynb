```r
library(dplyr)

dados <- read.csv("world_economics.csv")

dados <- dados %>%
  dplyr::select(name, currency, capital, languages, region, GDP, GDP.Growth,
                Interest.Rate, Inflation.Rate, Jobless.Rate, Gov..Budget,
                Debt.GDP, Current.Account, Population)

head(dados)
str(dados)
summary(dados)
```

```r
name <- factor(dados$name)
currency <- factor(dados$currency)
capital <- factor(dados$capital)
languages <- factor(dados$capital)
region <- factor(dados$region)
```

```r
cat("Médias das variavéis")
media_GDP <- mean(dados$GDP, na.rm = TRUE)
media_GDP.Growth <- mean(dados$GDP.Growth, na.rm = TRUE)
media_Interest.Rate <- mean(dados$Interest.Rate, na.rm = TRUE)
media_Inflation.Rate <- mean(dados$Inflation.Rate, na.rm = TRUE)
media_Jobless.Rate <- mean(dados$Jobless.Rate, na.rm = TRUE)
media_Gov..Budget <- mean(dados$Gov..Budget, na.rm = TRUE)
media_Debt.GDP <- mean(dados$Debt.GDP, na.rm = TRUE)
media_Current.Account <- mean(dados$Current.Account, na.rm = TRUE)
media_Population <- mean(dados$Population, na.rm = TRUE)

cat("\n\nGPD (média): ", media_GDP)
cat("\nGDP Growth (média): ", media_GDP.Growth)
cat("\nInterest Rate (média): ", media_Interest.Rate)
cat("\nInflation Rate (média): ", media_Inflation.Rate)
cat("\nJobless Rate (média): ", media_Jobless.Rate)
cat("\nGov. Budget (média): ", media_Gov..Budget)
cat("\nDebt/GDP (média): ", media_Debt.GDP)
cat("\nCurrent Account (média): ", media_Current.Account)
cat("\nPopulation (média): ", media_Population)

# Estatísticas descritivas para variáveis econômicas

cat("\n\nMediana das variáveis")
mediana_GDP <- median(dados$GDP, na.rm = TRUE)
mediana_GDP_Growth <- median(dados$GDP.Growth, na.rm = TRUE)
mediana_Interest <- median(dados$Interest.Rate, na.rm = TRUE)
mediana_Inflation <- median(dados$Inflation.Rate, na.rm = TRUE)
mediana_Jobless <- median(dados$Jobless.Rate, na.rm = TRUE)
mediana_Budget <- median(dados$Gov..Budget, na.rm = TRUE)
mediana_Debt <- median(dados$Debt.GDP, na.rm = TRUE)
mediana_Current <- median(dados$Current.Account, na.rm = TRUE)
mediana_Pop <- median(dados$Population, na.rm = TRUE)

cat("\n\nGDP (Mediana): ", mediana_GDP)
cat("\nGDP Growth (Mediana): ", mediana_GDP_Growth)
cat("\nInterest Rate (Mediana): ", mediana_Interest)
cat("\nInflation Rate (Mediana): ", mediana_Inflation)
cat("\nJobless Rate (Mediana): ", mediana_Jobless)
cat("\nGov. Budget (Mediana): ", mediana_Budget)
cat("\nDebt/GDP (Mediana): ", mediana_Debt)
cat("\nCurrent Account (Mediana): ", mediana_Current)
cat("\nPopulation (Mediana): ", mediana_Pop)


# Moda das variáveis
cat("\n\nModa das variáveis")
moda <- function(x) {
  x <- na.omit(x)                        # remove NAs
  ux <- unique(x)
  ux[which.max(tabulate(match(x, ux)))]
}

cat("\n\nGDP (Moda): ", moda(dados$GDP))
cat("\nGDP Growth (Moda): ", moda(dados$GDP.Growth))
cat("\nInterest Rate (Moda): ", moda(dados$Interest.Rate))
cat("\nInflation Rate (Moda): ", moda(dados$Inflation.Rate))
cat("\nJobless Rate (Moda): ", moda(dados$Jobless.Rate))
cat("\nGov. Budget (Moda): ", moda(dados$Gov..Budget))
cat("\nDebt/GDP (Moda): ", moda(dados$Debt.GDP))
cat("\nCurrent Account (Moda): ", moda(dados$Current.Account))
cat("\nPopulation (Moda): ", moda(dados$Population))


# Variância das variáveis
cat("\n\nVariância das variáveis")
var_GDP <- var(dados$GDP, na.rm = TRUE)
var_GDP_Growth <- var(dados$GDP.Growth, na.rm = TRUE)
var_Interest <- var(dados$Interest.Rate, na.rm = TRUE)
var_Inflation <- var(dados$Inflation.Rate, na.rm = TRUE)
var_Jobless <- var(dados$Jobless.Rate, na.rm = TRUE)
var_Budget <- var(dados$Gov..Budget, na.rm = TRUE)
var_Debt <- var(dados$Debt.GDP, na.rm = TRUE)
var_Current <- var(dados$Current.Account, na.rm = TRUE)
var_Pop <- var(dados$Population, na.rm = TRUE)

cat("\n\nGDP (Variância): ", var_GDP)
cat("\nGDP Growth (Variância): ", var_GDP_Growth)
cat("\nInterest Rate (Variância): ", var_Interest)
cat("\nInflation Rate (Variância): ", var_Inflation)
cat("\nJobless Rate (Variância): ", var_Jobless)
cat("\nGov. Budget (Variância): ", var_Budget)
cat("\nDebt/GDP (Variância): ", var_Debt)
cat("\nCurrent Account (Variância): ", var_Current)
cat("\nPopulation (Variância): ", var_Pop)


# Desvio Padrão das variáveis
cat("\n\nDesvio padrão das variáveis")
dp_GDP <- sd(dados$GDP, na.rm = TRUE)
dp_GDP_Growth <- sd(dados$GDP.Growth, na.rm = TRUE)
dp_Interest <- sd(dados$Interest.Rate, na.rm = TRUE)
dp_Inflation <- sd(dados$Inflation.Rate, na.rm = TRUE)
dp_Jobless <- sd(dados$Jobless.Rate, na.rm = TRUE)
dp_Budget <- sd(dados$Gov..Budget, na.rm = TRUE)
dp_Debt <- sd(dados$Debt.GDP, na.rm = TRUE)
dp_Current <- sd(dados$Current.Account, na.rm = TRUE)
dp_Pop <- sd(dados$Population, na.rm = TRUE)

cat("\n\nGDP (DP): ", dp_GDP)
cat("\nGDP Growth (DP): ", dp_GDP_Growth)
cat("\nInterest Rate (DP): ", dp_Interest)
cat("\nInflation Rate (DP): ", dp_Inflation)
cat("\nJobless Rate (DP): ", dp_Jobless)
cat("\nGov. Budget (DP): ", dp_Budget)
cat("\nDebt/GDP (DP): ", dp_Debt)
cat("\nCurrent Account (DP): ", dp_Current)
cat("\nPopulation (DP): ", dp_Pop)
```

```r
calcular_estatisticas <- function(df) {

  moda <- function(x) {
    x <- na.omit(x)
    ux <- unique(x)
    ux[which.max(tabulate(match(x, ux)))]
  }

  resultados <- lapply(df, function(x) {
    if (is.numeric(x)) {
      data.frame(
        Media = mean(x, na.rm = TRUE),
        Mediana = median(x, na.rm = TRUE),
        Moda = moda(x),
        Variancia = var(x, na.rm = TRUE),
        Desvio_Padrao = sd(x, na.rm = TRUE)
      )
    } else {
      NULL
    }
  })

  resultados <- do.call(rbind, resultados)
  resultados <- cbind(Variavel = names(df)[sapply(df, is.numeric)], resultados)

  rownames(resultados) <- NULL
  return(resultados)
}

estatisticas <- calcular_estatisticas(dados)
estatisticas
```

```r
freq_abs <- table(dados$region)
freq_rel <- prop.table(freq_abs) * 100

tabela_freq <- data.frame(
  Região = names(freq_abs),
  Frequência = as.vector(freq_abs),
  `Relativa(%)` = round(as.vector(freq_rel), 2))

print(tabela_freq)
```

```r

classes_pop <- cut(dados$Population,
                   breaks = c(0, 10, 50, 100, 500, 1500),
                   include.lowest = TRUE)

freq_abs <- table(classes_pop)
freq_rel <- prop.table(freq_abs) * 100

tabela_pop <- data.frame(
  Classe = names(freq_abs),
  Frequencia = as.vector(freq_abs),
  `Frequencia Relativa (%)` = round(as.vector(freq_rel), 2))

print(tabela_pop)

```

```r
freq_minima <- 2
freq_idiomas <- table(dados$languages)
idiomas_frequentes <- names(freq_idiomas[freq_idiomas >= freq_minima])
dados_filtrados <- dados[dados$languages %in% idiomas_frequentes, ]

tab_1 <- table(dados_filtrados$languages, dados_filtrados$region)
tab_freq_abs <- rbind(
    cbind(tab_1, margin.table(tab_1, 1)),
    c(margin.table(tab_1, 2), sum(margin.table(tab_1, 1)))
)

dimnames(tab_freq_abs)[[1]][nrow(tab_freq_abs)] <- "total linha"
dimnames(tab_freq_abs)[[2]][ncol(tab_freq_abs)] <- "total coluna"
kable(tab_freq_abs)
```
